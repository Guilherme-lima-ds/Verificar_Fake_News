```python
#Resetando os index para não da problema no futuro
df.reset_index(inplace=True, drop=True)
```

```python
#Verificando nossa coluna de texto, sabermos se vamos precisar realizar alguma transformação nela adicional.
df['text'].head(10)
```

- Como pode ser visto, na nossa coluna de texto existe, .,:' entre outros montes de caracteres que eu na hora do encondig por exemplo ou caso queira criar um WordCloud, não vou conseguir pois esses são caracteres especiais, que meu modelo não vai aceitar, então vamos limpa-los.

```python
#Limpando os caracteres especiais, numeros, etc...

#Variavel que remove todos os numeros junto com as letras.
remover_numeros = lambda x: re.sub('\w*\d\w*', ' ',x)

#Varaivel que transforma todos os valores em Lowercase(Letras misnusculas)
Lower_case = lambda x: re.sub('[%s]' % re.escape(string.punctuation), ' ', x.lower())

#Remover todos o '\n'(Tabs do teclado) e trocar por espaço
Remover_contra_barra = lambda x: re.sub('\n', ' ', x)

#Remover todos os non-ascii caracteres
remover_non_ascii = lambda x: re.sub(r'[^\x00-\x7f]',r' ',x)


#Aplicando todas as variaveis de uma vez nos meus dados.
df['text'] = df['text'].map(remover_numeros).map(Lower_case).map(Remover_contra_barra).map(remover_non_ascii)
```

```python
#Verificando
df['text'][5]
```

- Como podemos notar, não existe mais numeros,'. etc... tudo isso iria atrapalhar o desenvolvimento do projeto, então tivemos que excluir, agora com os textos limpos podemos seguir em frente na estruturação do nosso projeto.

```python
#Tirando valores indesejados
df['text'] = df['text'].replace('[^a-zA-Z]',' ',regex=True)

#Transformando as letras em miniculas
df['text'] = df['text'].str.lower()

#Removendo espaços adicionais
df['text']= df['text'].apply(lambda x: re.sub(' +',' ',x))
```

#### Etapa 5: Encondig dos dados
- Vamos usar dois tipos de encondig aqui
    - Encondig Vectorizado
        - O enconding vectorizado considera todas as palavras como um peso, ou seja, ele trás um score de cada palavra que aparecer, por exemplo, vamos dizer que ele divida em documentos e no documento 1 ele diga que a palavra people tem 0,70% de aparecer, ou seja, e um score que da a probabilidade da plavra estar lá, Resumindo: ele considera o peso de cada palavra.
    - Encondig Count Vectorizado
        - Já o Count vectorizado e algo mais preciso, e como se fosse uma variavel dummy, ele considera a frequência em relação ao indice do vocabulario, Resumindo: O count vectorizer fornece a frequência de cada palavra.

```python
#Para o começar o enconding eu vou separar o texto entre x e y
x = df['text']
y = df['label']
#Agora vou separar entre Treino e Teste
x_train, x_test, y_train, y_test = train_test_split(x,y, random_state=10, test_size=0.2)
```

- TfidfVectorizer

```python
#Usando o Tfidf para vetoriazar os dados
vector = TfidfVectorizer(stop_words='english', max_df=0.7)
#Fazendo o modelo aprender o peso de cada palavra.
vector_train = vector.fit_transform(x_train)
#Criando um novo x_test, so que dessa vez vectorizado pelo peso de cada palavra.
vector_test = vector.transform(x_test)
```

- Count Vectorizer

```python
#Count Vectorizer
count_vector = CountVectorizer(stop_words='english')
#Fazendo o modelo aprender a frequencia de cada palavra.
count_vector_train = count_vector.fit_transform(x_train.values)
#Criando um novo x_test, so que dessa vez vectorizado pela frequencia de cada palavra.
count_vector_test = count_vector.transform(x_test.values) 
```

#### Etapa 6: Aplicação do Modelo de ML e Avaliação dele.

```python
# Criando uma função que vai avaliar o modelo

def Avaliarmodelo(nome_modelo, y_test, previsao):
    sk = skplt.metrics.plot_confusion_matrix(y_test, previsao, normalize=True) #heatmap, vamos ver apenas para o modelo campeão.
    cm = classification_report(y_test, previsao, digits=4)
    ac = accuracy_score(y_test,previsao)
    roc = roc_auc_score(y_test, previsao)
    return print(f'O Modelo: {nome_modelo}\nRelatório de Classificação:\n{cm}\nAcurácia: {ac:.4f}\nAUC: {roc:.4f}\n')
```

```python
#Criando uma instancia para os modelos
modelo_rf = RandomForestClassifier()
modelo_lc = LogisticRegression()
modelo_xg = XGBClassifier(learning_rate=0.01, n_estimators=1000,
                         max_depth=3, subsample = 0.9,
                         colsample_bytree = 0.1, gamma=1,
                         random_state=42
)

#Criando um dicionario para acessar de maneira mais facil

modelos = {
    #'RandomForest Classifier':modelo_rf,
    'Logistic Regression': modelo_lc,
    #'XGB Classifier':modelo_xg
}
```

- Primeiro testando o Vectorizado

```python
#Percorrendo minha lista de modelos
for nome_modelo,Modelo in modelos.items():
    #Treinando modelo
    Modelo.fit(vector_train, y_train)
    
    #Prevendo com o modelo
    previsao = Modelo.predict(vector_test)
    
    #Avaliando o modelo
    Avaliarmodelo(nome_modelo,y_test,previsao)
    
    #Previsoes erradas.
    print('Previsões erradas sobre o Total')
    print((y_test !=previsao).sum(), '/',((y_test==previsao).sum()+(y_test !=previsao).sum()))
```

- Agora testando o Count Vectorizado

```python
#Percorrendo minha lista de modelos
for nome_modelo,Modelo in modelos.items():
    #Treinando modelo
    Modelo.fit(count_vector_train, y_train)
    
    #Prevendo com o modelo
    previsao = Modelo.predict(count_vector_test)
    
    #Avaliando o modelo
    Avaliarmodelo(nome_modelo,y_test,previsao)
    
    #Previsoes erradas.
    print('Previsões erradas sobre o Total')
    print((y_test !=previsao).sum(), '/',((y_test==previsao).sum()+(y_test !=previsao).sum()))
```
