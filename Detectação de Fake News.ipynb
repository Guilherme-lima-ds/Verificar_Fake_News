- Como podemos ver existem muitos valores desnecessarios como por exemplos ./' coisas que sõ vão atrabalhar o desenvolvimento do meu projeto, então vou precisar limpa-los.

```python
#Limpando texto função

def Limpartexto(texto):
    
    #Remover '\'
    texto = re.sub("'\'","", texto)
    
    #Remover simbolos especiais
    texto = re.sub('[^a-zA-Z]',' ',texto)
    
    #Remover pontuações
    texto = re.sub('[%s]' % re.escape(string.punctuation), ' ', texto)
    
    #Remover todos os non-ascii caracteres
    texto = re.sub(r'[^\x00-\x7f]',r' ',texto)
    
    #Remover todos o '\n'(Tabs do teclado) e trocar por espaço
    texto = re.sub('\n', ' ', texto)
     
    #Unindo textos com um espaço
    texto = ' '.join(texto.split())
    
    #Convertendo em Letras minisculas
    texto = texto.lower()
    
    return texto
```

```python
#Aplicando
df['text'] = df['text'].apply(lambda x:Limpartexto(x))
```

- Removendo StopWords
    - Stopwords são palavras ignoradas para fins de posicionamento em sites de pesquisa no Google. Alguns exemplos seriam artigos, preposições e pequenas palavras como: as, os, em, de, para, com, foi.

```python
#Baixando as stopowords
nltk.download('stopwords')

#Setando as palavras de stopwords
stop_words = set(stopwords.words('english'))

# Função que remove stopowords
def Removestopowords(texto):
    remover_stopowords = [word for word in texto.split() if word not in stop_words]
    return ' '.join(remover_stopowords)
```

Mantendo as palavras de Lematizing na minha senteça:
- Lemmatização é uma técnica de Processamento de Linguagem Natural que se propõe a reduzir uma palavra ao seu Lema , ou Forma Canônica.

```python
#Baixando as palavras de lematizing
nltk.download('wordnet')

#Instanciando
lemma = WordNetLemmatizer()

#Criando uma função que trás as palavras lematizing
def lematizing(sentece):
    stem_sentence= ""
    
    for word in sentece.split():
        stem = lemma.lemmatize(word)
        stem_sentence += stem
        stem_sentence += " "
    stem_sentence = stem_sentence.strip()
    return stem_sentence
```

```python
#Aplicando
df['text'] = df['text'].apply(lambda x:lematizing(x))
```

- Usando a tecnica de processamento Stemming
    - Qual e o processo de Stemming?
        - O processo de stemização (do inglês, stemming) consiste em reduzir uma palavra ao seu radical. A palavra “meninas” se reduziria a “menin”, assim como “meninos” e “menininhos”. As palavras “gato”, “gata”, “gatos” e “gatas” reduziriam-se para “gat”. A lematização reduz a palavra ao seu lema, que é a forma no masculino e singular.

```python
#instanciando a variavel
stemmer = PorterStemmer()

#Definindo a função.
def stemming(setence):
    stemmed_sentece = ""
    for word in setence.split():
        stem = stemmer.stem(word)
        stemmed_sentece += stem
        stemmed_sentece += " "
        
    stemmed_sentece = stemmed_sentece.strip()
    return stemmed_sentece
```

```python
#Aplicando
df['text'] = df['text'].apply(lambda x:stemming(x))
```

#### Etapa 5: Encondig dos dados
- Vamos usar dois tipos de encondig aqui
    - Encondig Vectorizado
        - O enconding vectorizado considera todas as palavras como um peso, ou seja, ele trás um score de cada palavra que aparecer, por exemplo, vamos dizer que ele divida em documentos, no documento 1 ele diga que a palavra people tem 0,70% de aparecer, ou seja, e um score que da a probabilidade da palavra estar lá, Resumindo: ele considera o peso de cada palavra.
    - Encondig Count Vectorizado
        - Já o Count vectorizado e algo mais preciso, e como se fosse uma variavel dummy, ele considera a frequência em relação ao indice do vocabulario, Resumindo: O count vectorizer fornece a frequência de cada palavra.

- TfidfVectorizer

```python
# TfidfVectorizer


#Usando o Label enconder para fazer o processo de encondig na minha variavel alvo Y
from sklearn.preprocessing import LabelEncoder

#Valor que vai ser codificado pelo tfiVector.
x = df['text']

#Instaciando o LabelEnconder
le = LabelEncoder()

#Transformando meu Y
y = le.fit_transform(df['label'])

#Separando entre treino e teste
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=10, test_size=0.2)
```

```python
#Usando o Tfidf para vetoriazar os dados
vector = TfidfVectorizer(stop_words='english', max_df=0.8)

#Fazendo o modelo aprender o peso de cada palavra.
vector_train = vector.fit_transform(x_train.values.astype('U'))

#Criando um novo x_test, so que dessa vez vectorizado pelo peso de cada palavra.
vector_test = vector.transform(x_test.values.astype('U'))
```

---

- Count Vectorizer

```python
#Fazendo com o count vectorizer, sem usar o label enconder dessa vez, vou colocar um "s" depois de cada variavel
# para diferenciar lá na frente os modelos de count e vec

#Para o começar o enconding eu vou separar o texto entre x e y
xs = df['text']
ys = df['label']
#Agora vou separar entre Treino e Teste
x_trains, x_tests, y_trains, y_tests = train_test_split(xs,ys, random_state=10, test_size=0.2)
```

```python
# Criando uma função que vai avaliar o modelo

def Avaliarmodelo(nome_modelo, y_test, previsao):
    sk = skplt.metrics.plot_confusion_matrix(y_test, previsao, normalize=True) #heatmap, vamos ver apenas para o modelo campeão.
    cm = classification_report(y_test, previsao, digits=4)
    ac = accuracy_score(y_test,previsao)
    roc = roc_auc_score(y_test, previsao)
    return print(f'O Modelo: {nome_modelo}\nRelatório de Classificação:\n{cm}\nAcurácia: {ac:.4f}\nAUC: {roc:.4f}\n')
```

```python
#Lista Vazia
generos_previstos = []

#Pecorrendo pelo tamanho da minha previsão
for i in range(len(ans)):
    #Pegando o index
    index_val = ans[i][0]
    #Fazendo o append de cada valor com seu determinado numero, 0,1,2
    generos_previstos.append(list(mapper.keys())[list(mapper.values()).index(index_val)])
```

```python
#Fazendo um dicionario com valor e chave para conseguimos fazer o append no nosso df
newmap = dict([(value,key)for key,value in mapper.items()])
newmap
```

```python
#Valores corretos, para cada valor que ele encontrar ele vai adicinar o nome fantasy, etc...
df['Noticia Correta'] = df['label'].map(newmap)
#Mesma coisa so que com valores previstos no meu array, e não mais valores definidos
df['Noticia Prevista'] = np.array(generos_previstos)
```
